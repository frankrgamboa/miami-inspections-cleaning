```python
# Creamos una copia para trabajar sin alterar el original
df_clean = df.copy()
# Convertir la columna 'date' a tipo datetime
df_clean['date'] = pd.to_datetime(df_clean['date'], errors='coerce')

# Crear columnas nuevas: mes, día de la semana y año
df_clean['mes'] = df_clean['date'].dt.month
df_clean['dia_semana'] = df_clean['date'].dt.day_name()
df_clean['anio'] = df_clean['date'].dt.year
columnas_a_eliminar = [
    'photo', 'super_email', 'super_phone',
    'dir_email', 'dir_phone', 'permit_sequence', 'inspector_id'
]

df_clean = df_clean.drop(columns=columnas_a_eliminar, errors='ignore')
df_clean = df_clean[
    df_clean['job_site_address'].notnull() &
    df_clean['inspection_description'].notnull() &
    df_clean['disp_description'].notnull()
]
# Convertimos a mayúsculas por consistencia
df_clean['disp_description'] = df_clean['disp_description'].str.upper()
df_clean['inspection_description'] = df_clean['inspection_description'].str.upper()
```

```python


fr = DF.drop(  ['date', 'request_date', 'anio', 'mes'], axis = 1)
```

```python
def clasificar_disp(desc):
    desc = str(desc).upper().strip()
    if ('APPROVED' in desc or 'APPROVE' in desc or 'APPRVD' in desc) and 'REJECTED' not in desc:
        return 'APPROVED'
    elif 'REJECTED' in desc or 'CORRECTION' in desc or 'NOT TO CODE' in desc:
        return 'REJECTED'
    elif 'CANCELLATION' in desc:
        return 'CANCELLATION'
    elif 'NOT READY' in desc or 'NOT PERFORMED' in desc:
        return 'NOT READY'
    else:
        return 'OTHER'

fr['result'] = fr['disp_description'].apply(clasificar_disp)

Df = fr.drop('disp_description', axis=1)

Df
```

```python
Df
```

```python

tabla_frec = Df.groupby(['permit_type', 'result']).size().unstack(fill_value=0)
pd.crosstab(Df['permit_type'], Df['result'], normalize='index').round(3)
```

```python
tabla_prop = tabla_frec.div(tabla_frec.sum(axis=1), axis=0)
display(tabla_prop)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.countplot(data=Df, x='permit_type', hue='result')
plt.title('Distribución de resultados por tipo de permiso')
plt.xlabel('Tipo de permiso')
plt.ylabel('Cantidad de inspecciones')
plt.xticks(rotation=45)
plt.legend(title='Resultado')
plt.tight_layout()
plt.show()
```

```python
tabla_prop = (
    Df.groupby(['permit_type', 'result'])
      .size()
      .groupby(level=0)
      .transform(lambda x: x / x.sum())
      .reset_index(name='proporcion')
)

plt.figure(figsize=(10, 6))
sns.barplot(
    data=tabla_prop,
    x='permit_type',
    y='proporcion',
    hue='result'
)

plt.title('Proporción de resultados por tipo de permiso')
plt.xlabel('Tipo de permiso')
plt.ylabel('Proporción')
plt.xticks(rotation=45)
plt.legend(title='Resultado')
plt.tight_layout()
plt.show()
```
